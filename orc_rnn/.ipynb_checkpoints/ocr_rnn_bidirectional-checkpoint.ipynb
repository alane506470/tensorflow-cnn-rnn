```python
import os
import gzip
import csv
import numpy as np
import tensorflow as tf
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
DOWNLOADED_FILENAME='letter.data'
def read_lines():
    with open(DOWNLOADED_FILENAME,'rt')as f:
        #\t表示空格隔開下一行
        reader= csv.reader(f,delimiter='\t')
        lines = list(reader)
        return lines
```

```python
lines=read_lines()
len(lines)
```

```python
#訓練數據集的特徵和標籤
def get_features_labels(lines):
    lines= sorted(lines,key=lambda x:int(x[0]))
    
    data,target=[],[]
    #新單詞的開頭
    next_id=-1
    #組成單詞的字元
    word=[]
    #images of those characters
    word_pixels=[]
    
    for line in lines:
        next_id=int(line[2])
        #print(next_id)
        pixels=np.array([int(x) for x in line[6:134]])
        pixels=pixels.reshape((16,8))
        
        word_pixels.append(pixels)
        word.append(line[1])
        
        if next_id ==-1:
            print(word)
        #當前單詞付加到data,target
            data.append(word_pixels)
            target.append(word)
            #清空
            word_pixels=[]
            word=[]
    return data,target
```

```python
data,target = get_features_labels(lines)
```

```python
#填充所有單詞，以便等於訓練數據中最常單詞的長度
def pad_feature_labels(data,target):
    max_length=max(len(x) for x in target)
    #print(max_length)
    padding = np.zeros((16,8))
    data = [o + [padding]*(max_length-len(o)) for o in data]
    target = [x + (['']*(max_length-len(x)))for x in target]
    return np.array(data), np.array(target)
```

```python
padded_data,padded_target = pad_feature_labels(data,target)
```

```python
reshaped_data=padded_data.reshape(padded_data.shape[:2]+(-1,))
```

```python
one_hot_target = np.zeros(padded_target.shape+(26,))
```

```python
#多維索引迭代器。
#返回一個迭代器，生成數組坐標和值對。
for index,letter in np.ndenumerate(padded_target):
    if letter:
        #print(index)
        #第一維索引可以照到特定單詞，第二為可以找到字母表示的字符的索引
        one_hot_target[index][ord(letter)-ord('a')]=1
        #print(one_hot_target[index])
```

```python
#隨機排序索引
shuffled_indices = np.random.permutation(len(reshaped_data))
print(shuffled_indices)
shuffled_data=reshaped_data[shuffled_indices]
shuffled_target= one_hot_target[shuffled_indices]
split = int(0.66*len(shuffled_data))

train_data=shuffled_data[:split]
train_target=shuffled_target[:split]

test_data=shuffled_data[split:]
test_target=shuffled_target[split:]
```

```python
#批量訓練財部會耗盡內存
#num_steps保存14張代表字符圖像(序列長度)
_,num_steps,num_inputs=train_data.shape
print(num_steps)
#保存輸出數量的變量
num_classes =train_target.shape[2]
```

```python
tf.reset_default_graph()

X=tf.placeholder(tf.float64,[None,num_steps,num_inputs])
y=tf.placeholder(tf.float64,[None,num_steps,num_classes])

#希望跟蹤訓練數據輸入的所有單詞的確切序列長度
#檢查圖像是否包含任何非0元素，e.g 'cat'  has 3 non zero elements
#可以得到所有批量的list包含14字符，字符有值則為1否則為0
used = tf.sign(tf.reduce_max(tf.abs(X),reduction_indices=2))
#計算所有批量的list有幾個1
length = tf.reduce_sum(used,reduction_indices=1)
sequence_length=tf.cast(length,tf.int32)
```

```python
num_neurons=300 
```

```python
#前向rnn，明確指定scope以免發生衝突
forward, _=tf.nn.dynamic_rnn(tf.nn.rnn_cell.GRUCell(num_neurons),X,dtype=tf.float64
                            ,sequence_length=sequence_length,scope='rnn-forward')
```

```python
#反轉輸入序列已匯入反向rnn
#反轉的維度是1及單詞中的字符
X_reversed=tf.reverse_sequence(X,sequence_length,seq_axis=1)
backward, _=tf.nn.dynamic_rnn(tf.nn.rnn_cell.GRUCell(num_neurons),X_reversed,dtype=tf.float64
                            ,sequence_length=sequence_length,scope='rnn-backward')
```

```python
#反轉後張量以原始順序返回輸出
backward = tf.reverse_sequence(backward,sequence_length,seq_axis=1)
```

```python
backward, forward
```

```python
output=tf.concat([forward,backward],axis=2)
```

```python
output.shape
```

```python
#透過設置權重和偏差來手動構建共享softmax曾，已乘上每個rnn層的輸出
#，然後應用softmax函數

#num_neurons是每一層神經元的數量，
weight = tf.Variable(tf.truncated_normal([num_neurons*2,num_classes],stddev=0.01,dtype=tf.float64))
bias = tf.Variable(tf.constant(0.1,shape=[num_classes],dtype=tf.float64))

#在應用權重前，將rnn的輸出攤平，rnn輸出的形狀(批量,14,300)
#，使用-1將錢2個此寸攤平，結果將是一個2維張量
flattened_output=tf.reshape(output,[-1,num_neurons*2])
#為softmax激活函數指定輸入，以概率形式獲得輸出
logits=tf.matmul(flattened_output,weight)+bias
#調整形狀使其與目標輸出張量相同(批量大小，14,26)
logits_reshaped=tf.reshape(logits,[-1,num_steps,num_classes])
#公共softmax有14個輸入，每個rnn層有1個，並產生14個輸出
#這些輸出的每一個都是預測字符的一件表示，一次性表示包含以下任意字符的輸出概率值

#自動將softmax激活應用於logits層
#還計算損失函數(交叉墒)，實際與預測的概率分布之差
cross_entropy=tf.nn.softmax_cross_entropy_with_logits(logits=logits,labels=y)
mistakes=tf.not_equal(tf.argmax(y,2),tf.argmax(logits_reshaped,2))

mistakes=tf.cast(mistakes,tf.float64)
#需要排除填充字符上的錯誤
#可以得到所有批量的list包含14字符，字符有值則為1否則為0
mask=tf.sign(tf.reduce_max(tf.abs(y),reduction_indices=2))
#如果標籤的一次性編碼的值為1則為實際值，如果不適則為填充
mistakes*=mask
#計算每個單詞的平均錯誤數
mistakes=tf.reduce_sum(mistakes,reduction_indices=1)
#除上實際序列長度
mistakes/=tf.cast(sequence_length,tf.float64)
#這是單詞中平均錯誤的數量
error = tf.reduce_mean(mistakes)
loss=tf.reduce_mean(cross_entropy)
optimizer=tf.train.RMSPropOptimizer(0.002)
gradient = optimizer.compute_gradients(loss)
optimize = optimizer.apply_gradients(gradient)

def batched(data,target, batch_size):
    epoch=0
    offset=0
    
    while True:
        old_offset=offset
        #對可用元素總數執行模運算來計算偏移量
        offset=(offset+batch_size)%(target.shape[0]-batch_size)
        
        if offset<old_offset:
            #隨機排序功能和標籤
            shuffled_indices=np.random.permutation(len(data))
            data = data[shuffled_indices]
            target =target[shuffled_indices]
            epoch+=1
        batch_data=data[offset:(offset+batch_size),:]
        batch_target = target[offset:(offset+batch_size),:]
        yield batch_data,batch_target,epoch
        
```

```python
batch_size=20
batches=batched(train_data,train_target,batch_size)
epochs=5
```

```python
with tf.Session() as sess:
    #初始化所有參數
    sess.run(tf.global_variables_initializer())
    for index,batch in enumerate(batches):
        batch_data = batch[0]
        batch_target = batch[1]
        
        epoch=batch[2]
        
        if epoch>=epochs:
            break
        
        feed={X:batch_data,y:batch_target}
        train_error,_=sess.run([error,optimize],feed)
        #對於每一批，在預測單詞時所犯錯誤的百分比
        print('{}: {:3.6}%'.format(index+1,100*train_error))
        
    test_feed={X:test_data,y:test_target}
    test_error,_=sess.run([error,optimize],test_feed)
    #對於每一批，在預測單詞時所犯錯誤的百分比
    print('Test error: {:3.6}%'.format(100*test_error))
```
